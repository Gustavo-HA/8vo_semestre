## Proyecto 1: Arithmetic Arranger  
#### **Completado**

```python
def arithmetic_arranger(lista, bool=False):
  import operator
  ops = {"+": operator.add, "-": operator.sub}
  if len(lista) > 5:
    return 'Error: Too many problems.'
  superior = ''
  inferior = ''
  guion = ''
  resultado = ''
  operador = ''
  for operacion in lista:
    miembros = operacion.split()
    if (len(miembros[0]) > 4 or len(miembros[2]) > 4):
      return 'Error: Numbers cannot be more than four digits.'
    elif (miembros[1] != '+' and miembros[1] != '-'):
      return "Error: Operator must be '+' or '-'."
    elif (not miembros[0].isnumeric() or not miembros[2].isnumeric()):
      return 'Error: Numbers must only contain digits.'
    superior = superior + ' ' + miembros[0]
    inferior = inferior + ' ' + miembros[2]
    resultado = resultado + ' ' + str(ops[miembros[1]](int(miembros[0]),
                                                       int(miembros[2])))
    operador = operador + ' ' + str(miembros[1])
    if len(miembros[0]) >= len(miembros[2]):
      guion = guion + ' ' + '-' * (len(miembros[0]) + 2)
    else:
      guion = guion + ' ' + '-' * (len(miembros[2]) + 2)

  arranged = ''

  for operacion in range(len(guion.split())):
    arranged = arranged + ' ' * (len(guion.split()[operacion]) - len(
      superior.split()[operacion])) + superior.split()[operacion] + ' ' * 4
  arranged = arranged[0:len(arranged)-4]
  arranged = arranged + '\n'

  for operacion in range(len(guion.split())):
    arranged = arranged + operador.split()[operacion] + ' ' * (
      len(guion.split()[operacion]) - len(inferior.split()[operacion]) -
      1) + inferior.split()[operacion] + ' ' * 4
  arranged = arranged[0:len(arranged)-4]
  arranged = arranged + '\n'

  for operacion in range(len(guion.split())):
    arranged = arranged + guion.split()[operacion] + ' ' * 4
  arranged = arranged[0:len(arranged)-4]

  if bool:
    arranged = arranged + '\n'
    for operacion in range(len(guion.split())):
      arranged = arranged + ' ' * (len(guion.split()[operacion]) - len(
        resultado.split()[operacion])) + resultado.split()[operacion] + ' ' * 4
    arranged = arranged[0:len(arranged) - 4]
    return arranged
  else:
    return arranged



print(arithmetic_arranger(["32 + 698", "3801 - 2", "45 + 43", "123 + 49"],True))
```

## Proyecto 2: Time Calculator  
#### **Completado**

```python
def add_time(start, duration, day = None):
    entrada = {'Hora' : int(start.split(':')[0]),
               'Minutos' : int(start.split(':')[1].split()[0]),
               'Meridiano' : start.split()[1]}
    duracion = {'Hora' : int(duration.split(':')[0]),
                'Minutos' : int(duration.split(':')[1])}
    salida = {'Hora' : None,
              'Minutos' : None,
              'Meridiano' : None,
              'DiaSemana' : None}
    semana = ('Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday')

    # Hora resultante
    if (duracion['Minutos']+entrada['Minutos'])>=60:
        salida['Hora'] = entrada['Hora']+duracion['Hora']+1
        salida['Minutos'] = entrada['Minutos'] + duracion['Minutos'] - 60
    else:
        salida['Hora'] = entrada['Hora']+duracion['Hora']
        salida['Minutos'] = entrada['Minutos'] + duracion['Minutos']
    ###

    # Días que han pasado
    dias = 0
    if entrada['Meridiano']=='AM':
        minleft = 1440
    else:
        minleft = 720
    
    if entrada['Hora']==12:
        minleft = minleft - entrada['Minutos']
    else:
        minleft = minleft - (entrada['Minutos'] + entrada['Hora']*60)
    
    if minleft <= (duracion['Hora']*60 + duracion['Minutos']):
        dias = 1 + int((duracion['Hora']*60 + duracion['Minutos']-minleft)/1440)
    #####
    
    # Resulta en PM o AM?
    if (duracion['Hora']*60 + duracion['Minutos']-minleft-(dias-1)*1440) < 720:
        salida['Meridiano'] = 'AM'
    else:
        salida['Meridiano'] = 'PM'

    # Dia de la semana
    if day != None:
        salida['DiaSemana'] = semana[int((1 + semana.index(day.capitalize())+dias) % 7)-1]
    ###
     
    # Formatting minutes and hours to print 'em
    salida['Minutos'] = str(salida['Minutos'])
    if int(salida['Minutos'])<10:
        salida['Minutos'] = '0'+salida['Minutos']

    salida['Hora'] = str(salida['Hora'] % 12)
    if salida['Hora']=='0':
        salida['Hora']='12'
    ###

    if day == None:
        if dias == 0:
            new_time = salida['Hora']+':'+salida['Minutos']+' '+salida['Meridiano']
        elif (dias == 1):
            new_time = salida['Hora']+':'+salida['Minutos']+' '+salida['Meridiano']+' (next day)'
        else:
            new_time = salida['Hora']+':'+salida['Minutos']+' '+salida['Meridiano']+f' ({dias} days later)'
    else:
        if dias == 0:
            new_time = salida['Hora']+':'+salida['Minutos']+' '+salida['Meridiano']+f", {salida['DiaSemana']}"
        elif (dias == 1):
            new_time = salida['Hora']+':'+salida['Minutos']+' '+salida['Meridiano']+f", {salida['DiaSemana']}"+' (next day)'
        else:
            new_time = salida['Hora']+':'+salida['Minutos']+' '+salida['Meridiano']+f", {salida['DiaSemana']}"+f' ({dias} days later)'

    return new_time


print(add_time("3:00 PM", "3:10"))
# Returns: 6:10 PM

print(add_time("11:30 AM", "2:32", "Monday"))
# Returns: 2:02 PM, Monday

print(add_time("11:43 AM", "00:20"))
# Returns: 12:03 PM

print(add_time("10:10 PM", "3:30"))
# Returns: 1:40 AM (next day)

print(add_time("11:43 PM", "24:20", "tueSday"))
# Returns: 12:03 AM, Thursday (2 days later)

print(add_time("6:30 PM", "205:12"))
# Returns: 7:42 AM (9 days later)
```

## Proyecto 3: Budget App    
#### **Completado**

```python
class Category:
  def __init__(self, cat):
    self.name = cat
    self.ledger = list()
    self.founds = 0
    
  def deposit(self, amount, description=""):
    self.ledger.append({"amount": amount, "description": description})
    self.founds += amount

  def withdraw(self, amount, description=""):
    if self.check_funds(amount):
      self.ledger.append({"amount": -1 * amount, "description": description})
      self.founds -= amount
    return self.check_funds(amount)

  def get_balance(self):
    return self.founds

  def transfer(self, amount, payee):
    if self.check_funds(amount):
      self.withdraw(amount, f"Transfer to {payee.name}")
      payee.deposit(amount, f"Transfer from {self.name}")
    return self.check_funds(amount)

  def check_funds(self, amount):
    return self.founds >= amount

  def __str__(self):
    string = self.name.center(30).replace(" ", "*") + '\n'
    for transaction in self.ledger:
      string += "{:<23}{:>7.2f}".format(transaction["description"][:23],
                                        transaction["amount"]) + "\n"
    string += 'Total: ' + str(self.founds)
    return string


def create_spend_chart(categories):
    barchart='Porcentaje gastado por categoria\n'
    totalspent = 0
    spentbycategory = dict()
    for category in categories:
        for movement in category.ledger:
            if movement["amount"]<0:
                totalspent -= movement["amount"]
                spentbycategory[category] = spentbycategory.get(category,0)-movement["amount"]

    for percentage in reversed(range(10+1)):
        percentage *= 10
        barchart += f"{percentage:>3}| "
        for category in categories:
            aux = int(spentbycategory[category]*10/totalspent)*10
            if aux >= percentage:
                barchart += 'o'+2*' '
            else:
                barchart += 3*' '
        barchart+='\n'
    barchart += 4*' '+(1+3*len(categories))*'-'+'\n'
    
    for i in range(max([len(category.name) for category in categories])):
        barchart += 5*' '
        for category in categories:
            try:
                barchart += category.name[i]+2*' '
            except:
                barchart += 3*' '
        barchart += '\n'
    barchart = barchart[:len(barchart)-2]+' '
    return barchart


food = Category("Food")
food.deposit(900, "deposit")
food.withdraw(10.15, "groceries")
food.withdraw(15.89, "restaurant and more food for dessert")
clothing = Category("Clothing")
food.transfer(50, clothing)
clothing.withdraw(25.55)
clothing.withdraw(100)
auto = Category("Auto")
auto.deposit(1000, "initial deposit")
auto.withdraw(15)
yo = Category("MiFoquita")
yo.deposit(14000,'PrimerSueldo')
yo.withdraw(500,'DiscoMel')
yo.withdraw(100,'PapasBote')
yo.withdraw(1000,'OsitoCariñosito')
yo.withdraw(10000,'Anillo')
yo.withdraw(300,'Condones')


print(create_spend_chart([food, clothing, auto, yo]))




```

## Proyecto 4: Polygon Area Calculator    
#### **Completado**

```python
class Rectangle:
  def __init__(self,width,height):
    self.width = width
    self.height = height

  def get_area(self):
    return self.width*self.height

  def get_perimeter(self):
    return 2*self.width + self.height*2

  def get_diagonal(self):
    return (self.width**2 + self.height**2)**0.5

  def get_picture(self):
    if self.width > 50 or self.height >50:
      return 'Too big for picture.'
    picture=''
    for line in range(self.height):
      picture+='*'*self.width+'\n'
    return picture

  def get_amount_inside(self,shape):
    return int(self.get_area()/shape.get_area())

  def set_width(self,width):
    self.width=width

  def set_height(self,height):
    self.height=height

  def __str__(self):
    string = "Rectangle(width={}, height={})".format(self.width,self.height)
    return string
  

class Square(Rectangle):
  def __init__(self,length):
    self.width = length
    self.height = length

  def set_width(self,length):
    self.width = length
    self.height = length

  def set_height(self,length):
    self.width = length
    self.height = length

  def set_side(self,length):
    self.width = length
    self.height = length

  def __str__(self):
    string = "Square(side={})".format(self.width)
    return string
  

rect = Rectangle(5, 10)
print(rect.get_area())
rect.set_width(3)
print(rect.get_perimeter())
print(rect)
print(rect.get_picture())

sq = Square(9)
print(sq.get_area())
sq.set_side(4)
print(sq.get_diagonal())
print(sq)
print(sq.get_picture())
```

# Proyecto 5: Probability Calculator

```python
import copy
import random
# Consider using the modules imported above.

class Hat:
  def __init__(self,**balls):
    self.balls = balls
    self.contents = [type[0] for type in balls.items() for i in range(type[1]) ]
    
  def draw(self, n):
    draws = list()
    if n > len(self.contents):
      draws = self.contents
      self.contents = list()
    else:
      draws = random.sample(self.contents,k=n)
      for removes in draws:
        self.contents.remove(removes)
    return draws


def experiment(hat, expected_balls, num_balls_drawn, num_experiments):
  M = 0
  aux = copy.copy(hat.contents)
  for i in range(num_experiments):
    condition = True
    drawns = hat.draw(num_balls_drawn)
    contador = dict()
    for drawn in drawns:
      contador[drawn] = contador.get(drawn,0)+1
    
    try:
      for expected in expected_balls.items():
        if contador[expected[0]] < expected[1]:
          condition = False
    except:
      condition = False
    
    if condition:
      M += 1
    hat.contents = copy.copy(aux)
  return float(M/num_experiments)

random.seed(95)
hat = Hat(blue=4, red=2, green=6)
probability = experiment(
    hat=hat,
    expected_balls={"blue": 2,
                    "red": 1},
    num_balls_drawn=4,
    num_experiments=1000)
print("Probability:", probability)
```
